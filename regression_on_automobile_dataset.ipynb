Min-Max Scaling (Normalization): Min-max scaling, also known as normalization, rescales the data to a specific range, typically between 0 and 1. It involves subtracting the minimum value of the feature and dividing by the range (maximum value minus minimum value). This method preserves the original distribution of the data but compresses it to the specified range. Min-max scaling is sensitive to outliers since the minimum and maximum values determine the scaling.

Standardization (Z-score Scaling): Standardization transforms the data to have a mean of 0 and a standard deviation of 1. It involves subtracting the mean of the feature and dividing by the standard deviation. Standardization centers the data around zero and brings it to a similar scale. This method preserves the shape of the distribution and is less sensitive to outliers compared to min-max scaling. Standardization is commonly used when the data is assumed to follow a Gaussian (normal) distribution.

Robust Scaling: Robust scaling is a normalization method that is less affected by outliers. It uses the median and interquartile range (IQR) instead of the mean and standard deviation. The data is subtracted by the median and divided by the IQR. Robust scaling is effective when the data contains significant outliers that can influence the mean and standard deviation in standardization. It preserves the shape of the distribution and is more robust to outliers.

Log Transformation: Log transformation involves taking the logarithm of the values. It can help normalize skewed distributions, especially when the data has a long tail on one side. Log transformation compresses large values more than small values, which can make the data more symmetric. This transformation can be useful when dealing with variables that have a multiplicative relationship with the target variable.

MaxAbsScaler is a normalization method that scales the data by dividing each feature by its maximum absolute value. It preserves the sign of the data, making it suitable for datasets where zero-centered scaling is not desired or appropriate.

Comparable Scaling: By normalizing the data, you bring all the features to a similar scale, ensuring that no single feature dominates the model's learning process due to its larger magnitude. This allows the model to give equal weight to each feature during the training process, resulting in more reliable and fair comparisons between the features.

Convergence Speed: Normalizing the data can improve the convergence speed of regression algorithms, particularly those that use iterative optimization techniques, such as gradient descent. With normalized data, the optimization process can converge more quickly since the algorithm does not have to deal with widely varying scales of the features. This can result in faster training times and more efficient model fitting.

Model Interpretation: Normalizing the data can facilitate the interpretation of the regression coefficients or feature importance. When the features are on the same scale, the coefficients can be directly compared to understand their relative importance in predicting the target variable.

Outlier Mitigation: Normalization can help in reducing the impact of outliers on the regression model.

```python
#applying feature selection
setup(data=automobile_dataset,target='price',
      feature_selection=True,feature_selection_method='classic')
cm=compare_models()
```

```python
#model performance using outlier performance
setup(data=automobile_dataset,target='price',
      remove_outliers=True,outliers_threshold=0.05)
cm=compare_models()
```

```python
#model performance using transformation
setup(data=automobile_dataset,target='price',
      transformation = True,transformation_method='yeo-johnson')
cm=compare_models()
```

Principal Component Analysis (PCA) is a dimensionality reduction technique used to transform high-dimensional data into a lower-dimensional space while retaining the most important information. It identifies a set of orthogonal axes, called principal components, that capture the maximum variance in the data.

PCA assumes linearity and may not be suitable for all types of data or relationships. It is primarily effective when the data exhibits a linear structure, and capturing variance is a primary objective.

By feature selection you reducing number of dimension by throwing out irrelevant information. By PCA, you reducing number of dimension by transforming to artificial set but retaining same information.

```python
#model performance using PCA
setup(data=automobile_dataset,target='price',
      pca=True,pca_method='linear')
cm=compare_models()
```

```python
#model performance using outier removal , normalization, transformation
setup(data=automobile_dataset,target= 'price',
      remove_outliers=True,outliers_threshold=0.05,
      normalize=True,normalize_method='zscore',
      transformation=True,transformation_method='yeo-johnson')
cm=compare_models()
```

```python
#BUILDING A SINGLE MODEL - RandomForest
from pycaret.datasets import get_data
from pycaret.regression import *
automobile_dataset=get_data("automobile")
setup(data=automobile_dataset,target='price')
rfmodel = create_model("rf")
```
